```python
import pandas as pd
```

```python
def is_number(x):
    try:
        float(x)
        return True 
    except:
        return False
```

```python
from ml24.src.constants import allowed_units
from model.utils import convert_to_std_units

df1 = pd.read_csv("ml24\\dataset\\train.csv")
df1["entity_list"] = df1.entity_value.str.split(n=1)



df1["entity_unit"] = df1.entity_list.map(lambda x: x[-1])



df1["entity_value"] = df1.entity_list.map(lambda x: x[0])



df1.drop("entity_list", axis="columns", inplace=True)



df1 = df1.loc[df1.entity_value.map(is_number)]
df1 = df1.loc[df1.entity_unit.isin(allowed_units)]
df1["entity_value"] = df1.entity_value.astype("float64")
df1["entity_value"] = df1.apply(
    lambda x: convert_to_std_units(
        x["entity_value"], x["entity_unit"], x["entity_name"]
    ),
    axis="columns",
)
```

```python
bad_values = set(df1.loc[(df1.entity_value < 1e-4) | (df1.entity_value > 1e4)].image_link.map(lambda x: x.split("/")[-1]))
```

```python
import os 
raw_dir = "images\\raw"
processed_dir = "images\\processed"
for img in bad_values:
    raw_path = os.path.join(raw_dir, img)
    processed_path = os.path.join(processed_dir, img)
    try:
        if os.path.exists(raw_path):
            os.remove(raw_path)
    except Exception as e:
        print(f"error with {raw_path}: {e}")
        continue
    try:
        os.remove(processed_path)
    except Exception as e:
        print(f"error with {processed_path}: {e}")
        continue
    print(f"removed {img}")
```
